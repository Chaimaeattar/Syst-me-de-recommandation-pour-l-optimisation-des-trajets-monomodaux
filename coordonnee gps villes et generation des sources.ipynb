```python
import pandas as pd
import os
import matplotlib.pyplot as plt
os.chdir ("C:\\Users\\Hp\\Desktop\\Chaimae\\stage\\PFE\\Dataset")
cwd=os.getcwd()
print(cwd)
df=pd.read_excel('latit.xlsx')
```

```python
df.head()
```

```python
#generer les destination à partir des sources
df1=df[df['Ville']=='TEMARA']
df1c=df[df['Ville']=='TEMARA']
df2=df[df['Ville']=='SALE']
df2c=df[df['Ville']=='SALE']
df3=df[df['Ville']=='casa']
df3c=df[df['Ville']=='casa']
df4=df[df['Ville']=='RABAT']
df4c=df[df['Ville']=='RABAT']
```

```python
result1 = pd.merge(df1,
                 df1c[["QUARTIER","Ville","lat","long"]],
                 on='Ville', 
                 how='right')
result2 = pd.merge(df2,
                 df2c[["QUARTIER","Ville"]],
                 on='Ville', 
                 how='right')
result3 = pd.merge(df3,
                 df3c[["QUARTIER","Ville"]],
                 on='Ville', 
                 how='right')
result4 = pd.merge(df4,
                 df4c[["QUARTIER","Ville"]],
                 on='Ville', 
                 how='right')
result1 = result1[result1['QUARTIER_x']!=result1['QUARTIER_y']]
result2 = result2[result2['QUARTIER_x']!=result2['QUARTIER_y']]
result3 = result3[result3['QUARTIER_x']!=result3['QUARTIER_y']]
result4 = result4[result4['QUARTIER_x']!=result4['QUARTIER_y']]
result = result1.append(result2)
result.append(result3)
result.append(result4)
result=result[['Ville','QUARTIER_y','lat_y','long_y','QUARTIER_x','lat_x','long_x']]
result.columns = ['VILLE', 'Q_src','lat','long','Q_dis','lat','long']
```

```python
export_excel = result.to_excel (r'C:\\Users\\Hp\\Desktop\\Chaimae\\stage\\PFE\\Dataset\TEMARA.xlsx', index = None, header=True)
```
