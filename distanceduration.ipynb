```python
import pandas as pd
import googlemaps
from itertools import tee
import os
import urllib.request, json
os.chdir ("C:\\Users\\Hp\\Desktop\\Chaimae\\stage\\PFE\\Dataset\\Covoiturage entre ville")
cwd=os.getcwd()
print(cwd)
df=pd.read_excel('finale.xlsx')
df['distance']=''
df['duration']=''
df.head()
```

```python
api_key = "AIzaSyB3arFbGSKP3zukzckd1Qb8YxWjV3Kb1ww"
base_url="https://maps.googleapis.com/maps/api/distancematrix/json?units=metric"

i=0

for index, row in df.iterrows():
    src = str(row['Q_src']).replace(' ',',')
    dis = str(row['Q_dis']).replace(' ',',')
    nav_request = '&origins={}&destinations={}&key={}'.format(src,dis,api_key)
    request = base_url + nav_request 
    json_result = json.loads(urllib.request.urlopen(request).read())
    try:
        for line in json_result['rows']:
            distance = line['elements'][0]['distance']['value'] 
            duration = line['elements'][0]['duration']['value']    
    except:
        distance = ''
        duration = ''
    df['duration'][i]=duration
   # print(type(duration))
    df['distance'][i]= distance
    i=i+1
```
