```python
import pandas as pd
import os
import matplotlib.pyplot as plt
os.chdir ("C:\\Users\\Hp\\Desktop\\Chaimae\\stage\\PFE\\Dataset\\dataset_final")
cwd=os.getcwd()
print(cwd)
df=pd.read_excel('cov_entre_ville.xlsx')
```

```python
#generer les dates et les heures de depart
from faker import Faker
from random import randint
df['date_depart']=''
df['heure_depart']=''
faker = Faker()
for i in range(849):
    df['date_depart'][i]=faker.date_between(start_date='-1y', end_date='-1y')
    df['heure_depart'][i]=""+str(randint(6,23))+":"+str(randint(0,59))+""
```

```python
from time import *
df['duree_conv']=''
```

```python
#convertir la durée en heure
for index,row in df.iterrows():
    df['duree_conv'].iloc[index]=strftime('%H:%M:%S', gmtime(df['Durée'].iloc[index]))
```

```python
def conv(x):
    seconds=x*60
    seconds = seconds % (24 * 3600) 
    hour = seconds // 3600
    seconds %= 3600
    minutes = seconds // 60
    
      
    return "%d:%02d" % (hour, minutes)
```

```python
df['duree_conv']=df['Durée'].apply(lambda x: conv(x))
```

```python
df.head()
```

```python
#concatener les dates de depart et les heures de depart
df['dt_dep']=''
df['dt_dep']=df['date_depart'].astype(str)+' '+df['heure_depart'].astype(str)
```

```python
df.dt_dep.head()
```

```python
from datetime import datetime 
from time import strptime
from datetime import timedelta
```

```python
#la durée des trajets pour les moyenes de transport ctm et autocar
df['duree_CTM']=df['Durée']*2
df['duree_autocar']=df['Durée']*3
```

```python
df['dt_dep'] = df['dt_dep'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M'))
```

```python
#generer les heures d'arrivees pour les moyens de transport
df['dt_arr_voiture'] = df['dt_dep']+df['Durée'].apply(lambda x: timedelta(minutes=x))
df['dt_arr_CTM'] = df['dt_dep']+df['duree_CTM'].apply(lambda x: timedelta(minutes=x))
df['dt_arr_autocar'] = df['dt_dep']+df['duree_autocar'].apply(lambda x: timedelta(minutes=x))
```

```python
#convertir en format strptime
df['dt_arr_voiture'] = df['dt_arr_voiture'].apply (lambda x: x.strftime("%d/%m/%Y, %H:%M"))
df['dt_arr_CTM'] = df['dt_arr_CTM'].apply (lambda x: x.strftime("%d/%m/%Y, %H:%M"))
df['dt_arr_autocar'] = df['dt_arr_autocar'].apply (lambda x: x.strftime("%d/%m/%Y, %H:%M"))
df['dt_dep'] = df['dt_dep'].apply (lambda x: x.strftime("%d/%m/%Y, %H:%M"))
```

```python
df=df.drop(['date_depart','heure_depart'], axis=1)
```

```python
#generer les prix à partir des distances
df["prix_covoiturage"]=df["Distance"]*0.2
df["prix_CTM"]=df["Distance"]*0.45
df["prix_autocar"]=df["Distance"]*0.30
```
