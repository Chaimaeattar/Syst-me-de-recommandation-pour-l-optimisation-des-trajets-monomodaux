```python
df['traffic']=''
l=['Friday','Saturday','Sunday']
l1=['RABAT','CASABLANCA','KENITRA','MEKNES','FES']
from random import randint
for index,row in df.iterrows():
    if (df['Ville de depart'].iloc[index] in l1 and df['weekday_name_dep'].iloc[index] in l):
        df['traffic'].iloc[index]=2
    elif(df['Ville de depart'].iloc[index] in l1 or df['weekday_name_dep'].iloc[index] in l):
        df['traffic'].iloc[index]=1
    elif(df['Distance'].iloc[index] < 100) :
        df['traffic'].iloc[index]=1
    else:
        df['traffic'].iloc[index]=randint(0,1)
```
