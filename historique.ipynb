```
hist = []
m = len(df)
for i in range(0,m):
    dataSample = df4.sample(1000,random_state=200)
    for j,k in zip(range(len(dataSample)),dataSample['id_trajet']):
      # print(j,k)
      hist.append([df['id_user'][i],dataSample['id_trajet'][k-1],dataSample['rating'][k-1]])
```

```
pd.DataFrame(hist).to_csv('/content/drive/My Drive/historiqueraing.csv',sep=';',header=["id_user","id_trajet",'rating'],index= False)
```
